```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np



netflix_df = pd.read_csv("C:/Users/shubh/Downloads/workspace/workspace/netflix_data.csv")
```

```python
ahead_of_1990 = netflix_df['release_year'] >= 1990
```

```python
bellow_of_1999 = netflix_df['release_year'] <= 1999
```

```python
netflix_df[np.logical_and(ahead_of_1990, bellow_of_1999)]
```

```python
the_90s_club = np.logical_and(ahead_of_1990, bellow_of_1999)
```

```python
duration = int(netflix_df[the_90s_club]['duration'].mode()[0])
```

```python
print(duration)
```

```python
action_movie_ = netflix_df['genre'].str.contains('action', case = False)
```

```python
club_90s_action = np.logical_and(the_90s_club, action_movie_)
```

```python
short_movie_count = len(netflix_df[np.logical_and(club_90s_action, netflix_df['duration'] < 90)])
```

```python
print(short_movie_count)
```
